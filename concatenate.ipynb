```python
# dependencies
import pandas as pd
```

```python
#pull in csv data
january_csv = "Data/january21-citibike-tripdata.csv"
february_csv = "Data/february21-citibike-tripdata.csv"
march_csv = "Data/march21-citibike-tripdata.csv"
```

```python
#create data frames
january_df = pd.read_csv(january_csv)
february_df = pd.read_csv(february_csv)
march_df = pd.read_csv(march_csv)
```

```python
# concatenate data frames together
data_frames = [january_df, february_df, march_df]
data = pd.concat(data_frames)
data
```

```python
data.dtypes
```

```python
#change gender to string
data['gender'] = data['gender'].astype('str')
```

```python
data.dtypes
```

```python
new_data = data["gender"].replace({"1": "male", "2": "female", "0":"unknown"}, inplace=True)
new_data
```

```python
#export new data frame as CSV
data.to_csv("newdata.csv", index = False, header=True)
```
